```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.cifar10.load_data()
```

```python
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer','dog', 'frog', 'horse', 'ship', 'truck']
```

```python
labels = ['cat', 'dog', 'horse']
```

```python
X_train = []
Y_train = []
label_train = []

for i in range(len(train_images)):
  if train_labels[i] in [3, 5, 7]:
    if train_labels[i] == 3:
      label_train = 0
    elif train_labels[i] == 5:
      label_train = 1
    elif train_labels[i] == 7:
      label_train = 2
  
    X_train.append(train_images[i])
    Y_train.append([label_train])

X_train = np.array(X_train)
Y_train = np.array(Y_train)
```

```python
print("Total training examples: " + str(len(X_train)))
```

```python
plt.figure(figsize=(10,10))
for i in range(12):
    plt.subplot(4,3,i+1)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(X_train[i])
    plt.xlabel(labels[Y_train[i][0]])

plt.show()
```

```python
np.random.seed(1)
mu = 0
sigma = 1

W = np.array(np.random.normal(mu, sigma, 3*3072)).reshape(3,3072)
b = np.array(np.random.normal(mu, sigma, 3*1)).reshape(3,1)
```

```python
predictions = []
for i in range(X_train.shape[0]):
  x = X_train[i].flatten().reshape(3072,1)
  pred = np.dot(W,x) + b;
  idx = np.argmax(pred)
  predictions.append([idx])
  
```

```python
plt.figure(figsize=(10,10))
for i in range(12):
    plt.subplot(4,3,i+1)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(X_train[i])
    plt.xlabel(labels[predictions[i][0]])
plt.show()
```

```python
correct_predictions = 0
accuracy = 0

for i in range(len(predictions)):
  if predictions[i][0] == Y_train[i][0]:
    correct_predictions += 1

accuracy = (correct_predictions / len(predictions)) * 100

print("Accuracy = " + str(accuracy) + "%")
```
